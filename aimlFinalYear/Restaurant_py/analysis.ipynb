```python
d1={}
for i in range(len(df['items'].unique())):
    print(df['items'][i])
    d1[df['items'][i]]=i
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
```

```python
X=vectorizer.fit_transform(df['items'])
```

```python
X
```

```python
df_new=df
df_new.replace(d1,inplace=True)
```

```python
df_new
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(3)
kmeans.fit(df_new)
```

```python
kmeans
```

```python
identified_clusters = kmeans.fit_predict(X)
identified_clusters
```

```python
import numpy as np
np.unique(identified_clusters)
```

```python
np.unique(identified_clusters, return_counts=True)
```

```python
pos1=[]
pos2=[]
pos3=[]
for i  in range(len(identified_clusters)):
    if identified_clusters[i]==0:
        pos1.append(i)
    elif identified_clusters[i]==1:
        pos2.append(i)
    elif identified_clusters[i]==2:
        pos3.append(i)
        
```
