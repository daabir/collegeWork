```python
def save_details(item, date,time):
    sql = "INSERT INTO item_usage_details (item, rdate,rtime) VALUES (%s, %s,%s)"
    val = (item, date,time)
    mycursor.execute(sql, val)
    mydb.commit()

    print(mycursor.rowcount, "record inserted.")
```
